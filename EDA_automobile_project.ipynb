# Exploratory Data Analysis (EDA) 

```python
data.columns
```

```python
# How many unique car makes are there in the dataset? List the top 10 most common car makes.
```

```python
unique_cars=data["make"].unique()
print("ALL Unique Cars:\n")
print(unique_cars,"\n")

unique_cars_makes=data["make"].nunique()
print("Total Unique Cars Makes:",unique_cars_makes,"\n")

top_10_cars=data["make"].value_counts(ascending=False).head(10)
print("Top 10 Most Common Cars :\n")
print(top_10_cars)
```

```python
#piechart

plt.pie(top_10_cars, labels=top_10_cars.index, autopct='%1.1f%%',shadow = True)
plt.title('Top 10 Most Common Car Makes', fontsize=16)
plt.legend(title="Top 10 Most Common Car Makes",loc="upper left",bbox_to_anchor=(1.20,1),facecolor = 'yellow',shadow=True)
plt.grid()
plt.show()
```

```python
# Create a bar chart showing the distribution of car body styles.


ax=sns.countplot(x="body-style",data=data,saturation=1)
ax.bar_label(ax.containers[0])
plt.title("Distribution of Car Body Style")
plt.grid()
plt.show()
```

```python
# Explore the frequency of different fuel types using a pie chart or bar chart.

fuel_type_counts=data["fuel-type"].value_counts()

plt.pie(fuel_type_counts, labels=fuel_type_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen'],shadow=True)
plt.title("Frequency of Different Fuel Types",fontsize=15)
plt.legend(title="Frequency of Different Fuel Types",labels=fuel_type_counts.index,bbox_to_anchor=(1.20,1),shadow=True)
plt.show()
```

```python
# Investigate the relationship between engine size and horsepower using a scatter plot.
plt.figure(figsize=(8,6))
sns.scatterplot(x="engine-size",y="horsepower",data=data,hue="make")
plt.title("Relationship Between Engine Size and Horsepower",fontsize=15)
plt.legend(title="make",bbox_to_anchor=(1.20,1),shadow=True)
plt.show()
```
